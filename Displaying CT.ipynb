```python
import dicom
import numpy as np
from glob import glob
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
```

```python
filenames = [
    'CT1.2.840.113619.2.278.3.380434001.422.1404078961.991.{}.dcm'.format(i)
    for i in range(1,129)
]
filenames
```

```python
ct_data = np.empty((512,512,0))

for filename in filenames:
    dcm = dicom.read_file(filename, force=True)
    ct_data = np.concatenate((ct_data, dcm.pixel_array[:,:,None]), axis=2)
```

```python
np.shape(ct_data)
```

```python
y = x = np.arange(512)
```

```python
for i in np.arange(0,128,10):
    plt.contourf(
        x, y, ct_data[:, :, i], 30, 
        vmin=ct_data.min(), vmax=ct_data.max(), cmap=plt.get_cmap('gray'))
    plt.colorbar()
    plt.axis("equal")
    plt.show()
```
